# Data vizualization with Matplotlib and Seaborn libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
```

```python
df = pd.read_csv('./Data/tmp.txt', sep = ';', header = None)
df.head()
```

```python
df.columns = ['station_id', 'year', 'jan', 'feb', 'mar', 'apr', 'may', 'june', 'july', 'aug', 'sept', 'oct', 'nov', 'dec']
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
months = df.columns[2:]
```

```python
df[months] = df[months].apply(pd.to_numeric, errors='coerce')
```

```python
msno.matrix(df);
```

```python
msno.bar(df);
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['station_id'].unique()
```

```python
station1 = df[df['station_id'] == 20046].set_index('year')['feb']
station1
```

```python
plt.title('   20046    ')
plt.xlabel('')
plt.ylabel('')
plt.plot(station1)
plt.grid(True)
plt.axvline(x=2005, color='r', linestyle='--')
plt.axhline(y=-13.5, color='r', linestyle='--')
plt.show()
```

```python
station2 = df[df['station_id'] == 20046].set_index('year')[df.columns[2:]]
station2.head()
```

```python
station2.shape
```

```python
station = station2.groupby('year')[station2.columns[2:]].mean()
```

```python
station.head()
```

```python
station2_mean = round(station2.mean(axis=1), 2)
```

```python
station2_mean.head()
```

```python
plt.title('   20046    ')
plt.xlabel('')
plt.ylabel('')
plt.plot(station2_mean)
plt.show()
```

```python
station2_mean.index.min()
```
