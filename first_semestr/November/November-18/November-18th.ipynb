# Data vizualization with Matplotlib and Seaborn libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Data/tmp.txt', sep = ';', header = None)
df.head()
```

```python
df.columns = ['station_id', 'year', 'jan', 'feb', 'mar', 'apr', 'may', 'june', 'july', 'aug', 'sept', 'oct', 'nov', 'dec']
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
months = df.columns[2:]
```

```python
df[months] = df[months].apply(pd.to_numeric, errors='coerce')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

# Вывод графика в цикле

```python
months_list = list(months)
print(months_list)
```

```python
for month in months_list:
    data = df[df['station_id'] == 20046][['year', month]]
    
    fig, ax = plt.subplots(figsize=(16, 9))
    ax.plot(data['year'], data[month], label = str(month))
    ax.set_title('Температура станции 20046 в феврале', pad=25, fontsize=25, fontfamily='Times New Roman')
    ax.set_xlabel('Год', fontsize = 15, fontfamily='Times New Roman')
    ax.set_ylabel('Температура', fontsize = 15, fontfamily='Times New Roman')
    ax.set_xticks(list(range(data['year'].min(), data['year'].max(), 10)) + [data['year'].max()])
```

# Вывод нескольких графиков на одном полотне

```python
fig, axs = plt.subplots(figsize=(16, 9), ncols = 1, nrows = 2, sharex = True)
data1 = df[df['station_id'] == 20046]['feb']
data2 = df[df['station_id'] == 26997]['feb']

axs[0].hist(data1)
axs[0].set_title('Гистограмма температуры для станции 20046 в феврале')

axs[1].hist(data2)
axs[1].set_title('Гистограмма температуры для станции 26997 в феврале')
axs[1].set_xlabel('Температура', fontsize = 15, fontfamily='Times New Roman');
```

```python
stations = [20046, 26997]
```

```python
fig, axs = plt.subplots(figsize=(15, 30), ncols = 2, nrows = 12, sharey = True)

for i, month in enumerate(months):
    for j, station in enumerate(stations):
        data = df[df['station_id'] == station][month]
        axs[i][j].hist(data)
        ax.set_xlabel('Температура', fontsize = 15, fontfamily='Times New Roman')
        
fig.suptitle('Температура для станцний 20046 и 26997 по месяцам', y = .9)
plt.show()
```

# Библиотека Seaborn

```python
import seaborn as sns
```

```python
data = df[df['station_id'] == 20046]['feb']
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
sns.histplot(data, ax=ax, kde=True)
ax.set_title('Гистограмма распределения температуры для станции 20046', fontsize = 25, fontfamily='Times New Roman', pad=25, fontweight='bold')
ax.set_xlabel('Температура', fontsize = 15, fontfamily='Times New Roman')
sns.set_style('whitegrid');
```

```python
data = df[df['station_id'] == 20046][['feb', 'mar']]
fig, ax = plt.subplots(figsize=(16, 9))
sns.histplot(data, ax=ax, alpha = .25)
ax.set_title('Гистограмма распределения температуры для станции 20046 в февралеи марте', fontsize = 25, fontfamily='Times New Roman')
ax.set_xlabel('Температура');
```

```python
station = df[df['station_id'] == 20046].set_index('year')[['feb', 'mar']]
fig, ax = plt.subplots(figsize=(16, 9))
sns.lineplot(data=station, ax=ax)
ax.set_title('Температура для станции 20046 в феврале и марте по годам', fontsize = 25, fontfamily='Times New Roman', fontweight='bold')
ax.set_xlabel('Год')
ax.set_ylabel('Температура');
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
data = df[df['station_id'] == 20046]['jan']
sns.scatterplot(data=data)

ax.set_title('Температура для станции 20046 в апреле', fontsize = 25, fontfamily='Times New Roman', fontweight='bold', pad=25)
ax.set_ylabel('Температура');
```
