**Чтение и запись данных средствами Python**

```python
my_text = 'Это просто текст'
```

```python
f = open('new_file.txt','w')
f.write(my_text)
f.close()
```

```python
f = open('new_file.txt','r')
data = f.read()
f.close()
data
```

```python
#Менеджер контекста
with open ('file_txt', 'w') as f:
  f.write('Это новый текст')
```

```python
with open('file_txt', 'r') as file:
  data = file.read()
```
