**Кодирование данных средствами SKlearn**

Label Encoding

```python
data = ['Paris', 'London', 'Moscow']
```

```python
le = preprocessing.LabelEncoder()
le.fit(data)
```

```python
le.classes_
```

```python
data_coded=le.transform(data)
data_coded
```

```python
le.inverse_transform(data_coded)
```

One-hot encoding

```python
targets = np.array(['red','red','green','blue','yellow','pink','white'])
```

```python
onehot_enc = preprocessing.OneHotEncoder()
encoded = onehot_enc.fit_transform(targets.reshape(-1,1))
encoded
```
