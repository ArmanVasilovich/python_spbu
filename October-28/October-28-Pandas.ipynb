# Подгружаем csv файл

```python
df = pd.read_csv('WeightLoss.csv')
df
```

```python
df.head()
```

```python
df['total'] = df['w1'] + df['w2'] + df['w3']
```

```python
df['total_gr'] = df['total']*1000
df.head()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df['total_gr_log'] = df['total_gr'].apply(np.log)
df.describe().T
```

```python
df['av_loss'] = df.loc[:,'w1' : 'w3'].apply(np.mean, axis = 1)
df.head()
```

```python
my_func = lambda x: x.max() - x.min()
```

```python
df['wrange'] = df.loc[:,'w1' : 'w3'].apply(my_func, axis = 1)
df.head()
```

```python
ages = pd.DataFrame({'age': [24, 25, np.nan, 29],
                     'income': [20000, np.nan, 26000, 30000],
                     'children': [2,1,3, np.nan]})
ages
```

```python
ages.mean()
```

```python
ages.fillna(ages.mean())
```

```python
# Заполняет None значениями выше
ages.fillna(method = 'ffill')
```

```python
# Заполняет None значениями ниже
ages.fillna(method = 'bfill')
```
